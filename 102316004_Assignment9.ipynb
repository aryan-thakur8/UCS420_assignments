```python
import nltk
import string
from nltk.corpus import stopwords
from collections import Counter
from nltk.tokenize import sent_tokenize,word_tokenize
import matplotlib.pyplot as plt
```

```python
nltk.download("all")
```

```python
text = """technology has drastically changed the way we live work and communicate the advancement of smartphones artificial intelligence and the internet of things has connected the world in ways never seen before people are now able to access information instantly and communicate with anyone anywhere this has opened up new opportunities in education business and social interaction but it also brings challenges such as privacy concerns and dependency on technology"""
```

Q1. Write a unique paragraph (5-6 sentences) about your favorite topic (e.g., sports, technology, food, books, etc.).

1.Convert text to lowercase and remove punctuaƟon.
2.Tokenize the text into words and sentences.
3.Remove stopwords (using NLTK's stopwords list).
4.Display word frequency distribuƟon (excluding stopwords).

```python
# Convert text to lowercase and remove punctuation
text = text.lower()
text_clean = text.translate(str.maketrans('', '', string.punctuation))
print("Cleaned Text:")
print(text_clean)

# Tokenize the text into words
words = nltk.word_tokenize(text_clean)
sentences = nltk.sent_tokenize(text)
print("\nTokens:")
print(words)
print("\nSentences:")
print(sentences)

# Remove stopwords using NLTK's stopwords list
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]
print("\nFiltered Words:")
print(filtered_words)

# Display word frequency distribution
word_freq = Counter(filtered_words)

# Plot the word frequency distribution
plt.figure(figsize=(10,6))
plt.bar(word_freq.keys(), word_freq.values())
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Word Frequency Distribution (Excluding Stopwords)')
plt.xticks(rotation=90)

plt.show()
print(word_freq)
```

Q2: Stemming and Lemmatization

1.Take the tokenized words from Question 1 (aŌer stopword removal).
2.Apply stemming using NLTK's PorterStemmer and LancasterStemmer.
3.Apply lemmatizion using NLTK's WordNetLemmaƟzer.
4.Compare and display results of both techniques.

```python
from nltk.stem import PorterStemmer,LancasterStemmer
from nltk.stem import WordNetLemmatizer
from collections import Counter
```

```python
# Initialize stemmers and lemmatizer
porter_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer()
lemmatizer = WordNetLemmatizer()

# Apply stemming using PorterStemmer and LancasterStemmer
porter_stemmed = [porter_stemmer.stem(word) for word in filtered_words]
lancaster_stemmed = [lancaster_stemmer.stem(word) for word in filtered_words]

# Apply lemmatization using WordNetLemmatizer
lemmatized = [lemmatizer.lemmatize(word) for word in filtered_words]

# Display results
print("Original Words: ", filtered_words)
print("\nPorter Stemmed Words: ", porter_stemmed)
print("\nLancaster Stemmed Words: ", lancaster_stemmed)
print("\nLemmatized Words: ", lemmatized)
```

```python
def custom_tokenizer(text):
    text = text.lower()
    pattern = r"\b(?:\d+\.\d+|\d+|[a-zA-Z]+(?:'[a-z]+)?(?:-[a-z]+)*)\b"
    tokens = re.findall(pattern, text)
    return tokens

custom_tokens = custom_tokenizer(text)
print("Custom Tokens:\n", custom_tokens)

text = re.sub(r'\b[\w.-]+@[\w.-]+\.\w+\b', '<EMAIL>', text)
text = re.sub(r'https?://\S+|www\.\S+', '<URL>', text)
text = re.sub(r'\+?\d{1,3}[ -]?\d{10}|\d{3}-\d{3}-\d{4}', '<PHONE>', text)

print("Cleaned Text:\n", text)
```
